```python
import pandas as pd
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

concatenated_tags = []
for tags in df['tags'].tolist():
    if isinstance(tags, list):
        concatenated_tags.extend(tags)
    elif not pd.isna(tags):  
        concatenated_tags.append(tags)
        
for tags in temp2['tags'].tolist():
    if isinstance(tags, list):
        concatenated_tags.extend(tags)
    elif not pd.isna(tags):  
        concatenated_tags.append(tags)

max_features = 500
tokenizer = Tokenizer(num_words=max_features)
tokenizer.fit_on_texts(concatenated_tags)

encoded_sequences = tokenizer.texts_to_sequences(concatenated_tags)

max_length = max(len(seq) for seq in encoded_sequences)
padded_sequences = pad_sequences(encoded_sequences, maxlen=max_length, padding='post')

df1_len = len(df)
df2_len = len(temp2)
df['encoded_sequence'] = list(padded_sequences[:df1_len])
temp2['encoded_sequence'] = list(padded_sequences[df1_len:df1_len+df2_len])

df.drop(columns=['tags'], inplace=True)
temp2.drop(columns=['tags'], inplace=True)

print(df)
print(temp2)
```

```python
import random
new_df = pd.DataFrame(index=range(len_df), columns=range(len_temp2), dtype=int)
for i in range(len_df):
    similarities = (q1_df.iloc[i] @ q2_df.T) / (np.linalg.norm(q1_df.iloc[i]) * np.linalg.norm(q2_df, axis=1))
    nearest_indices = similarities.argsort()[-50:][::-1]
    selected_indices = random.sample(nearest_indices.tolist(), 20)
    new_df.iloc[i, selected_indices] = 1
```

## Non-Matrix Factorization

```python
# https://ieeexplore.ieee.org/document/6165290/

# rESEARCH pAPER FOR nmf/
```

```python
from sklearn.decomposition import NMF

nmf_model = NMF(n_components=10, random_state=42)  

nmf_features = nmf_model.fit_transform(new_df)

W = nmf_model.components_
H = nmf_features
```

```python
W_df = pd.DataFrame(W, columns=new_df.columns)
H_df = pd.DataFrame(H, index=new_df.index)
```

```python
W_df.head()
```

```python
H_df.head()
```

```python
def content_based_recommendations(question_id, num_recommendations=5):
    similarities = W_df.dot(W_df.loc[question_id])
    similar_questions = similarities.sort_values(ascending=False).index[1:num_recommendations+1]
    return similar_questions

def collaborative_filtering_recommendations(question_id, num_recommendations=5):
    similarities = H_df.dot(H_df.loc[question_id])
    similar_questions = similarities.sort_values(ascending=False).index[1:num_recommendations+1]
    return similar_questions

exhibitor_id = 5
content_based_recs = content_based_recommendations(exhibitor_id)
collaborative_filtering_recs = collaborative_filtering_recommendations(exhibitor_id)

print("Content-Based Recommendations:", content_based_recs)
print("Collaborative Filtering Recommendations:", collaborative_filtering_recs)
```
